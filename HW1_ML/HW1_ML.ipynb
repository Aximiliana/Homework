```python
#5 Посмотрим как росло количество фильмов и их бюджет по годам
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
released_years_mov = mov_by_votes.groupby('year_release')[['title']].count()
released_years_mov.head()
```

```python
budget_movs = mov_by_votes.groupby('year_release')[['budget']].sum()
budget_movs.head()
```

```python
years_movs = released_years_mov.join(budget_movs)
years_movs.head()
```

```python
years_movs.columns = ['Number_of_movies', 'Budget_movies']
years_movs.head()
```

```python
trace0 = go.Scatter(
    x=years_movs.index,
    y=years_movs.Number_of_movies,
    name='Number_of_movies'
)

trace1 = go.Scatter(
    x=years_movs.index,
    y=years_movs.Budget_movies,
    name='Budget_movies'
)

data = [trace0, trace1]
layout = {'title': 'Statistics of movies'}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```
